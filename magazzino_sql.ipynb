```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')

# Creiamo una tabella prodotti magazzino
conn.execute('''
    CREATE TABLE magazzino (
        id INTEGER PRIMARY KEY,
        prodotto TEXT,
        quantita INTEGER,
        prezzo REAL,
        categoria TEXT
    )
''')

# Inseriamo prodotti
conn.execute("INSERT INTO magazzino VALUES (1, 'Mele', 150, 1.20, 'Frutta')")
conn.execute("INSERT INTO magazzino VALUES (2, 'Banane', 80, 0.90, 'Frutta')")
conn.execute("INSERT INTO magazzino VALUES (3, 'Patate', 200, 0.50, 'Verdura')")
conn.execute("INSERT INTO magazzino VALUES (4, 'Pomodori', 120, 1.50, 'Verdura')")
conn.execute("INSERT INTO magazzino VALUES (5, 'Arance', 60, 1.80, 'Frutta')")
conn.execute("INSERT INTO magazzino VALUES (6, 'Zucchine', 90, 0.80, 'Verdura')")
conn.execute("INSERT INTO magazzino VALUES (7, 'Fragole', 40, 3.50, 'Frutta')")

conn.commit()
print("Magazzino creato!")
```

```python
# Vediamo tutto il magazzino
df = pd.read_sql_query("SELECT * FROM magazzino", conn)
print(df)
```

```python
# Valore totale per ogni prodotto
df = pd.read_sql_query("""
    SELECT
        prodotto,
        quantita,
        prezzo,
        quantita * prezzo as valore_totale
    FROM magazzino
    ORDER BY valore_totale DESC
""", conn)
print(df)
```

```python
# Totale valore per categoria
df = pd.read_sql_query("""
    SELECT
        categoria,
        COUNT(*) as num_prodotti,
        SUM(quantita) as quantita_totale,
        ROUND(SUM(quantita * prezzo), 2) as valore_totale
    FROM magazzino
    GROUP BY categoria
    ORDER BY valore_totale DESC
""", conn)
print(df)
```

```python
# Prodotti con scorta bassa
df = pd.read_sql_query("""
    SELECT prodotto, quantita, categoria
    FROM magazzino
    WHERE quantita < 70
    ORDER BY quantita ASC
""", conn)
print(df)
```

Ora un esercizio da solo. Scrivi una query che mostra solo i prodotti con prezzo maggiore di 1€, ordinati per prezzo dal più caro al più economico.

```python
# Prodotti con prezzo maggiore di 1€, dal più caro al più economico
df = pd.read_sql_query("""
    SELECT
        prodotto,
        prezzo
    FROM magazzino
    WHERE prezzo > 1.0
    ORDER BY prezzo DESC
""", conn)
print(df)
```

Ultimo esercizio da solo — scrivi una query che mostra la media del prezzo per categoria.


```python
df = pd.read_sql_query("""
    SELECT
        categoria,
        AVG(prezzo) as prezzo_medio
    FROM magazzino
    GROUP BY categoria
""", conn)
print(df)
```
